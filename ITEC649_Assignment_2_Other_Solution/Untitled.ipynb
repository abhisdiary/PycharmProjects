```python
import csv
from bs4 import BeautifulSoup
with open("/home/shreyas_ashwath/Desktop/a2-master/index.html") as fd:
    html_doc=fd.read()
soup=BeautifulSoup(html_doc,"html.parser")
```

```python
productDivs = soup.findAll('div', attrs={'class' : 'product'})
cost,product_ids,product_name,stock=list(),list(),list(),list()
product_info=dict()
for div in productDivs:
    cost.append(div.find("div", {"class": "cost"}).string) #cost
    product_ids.append(div.find("a")["href"].split("/")[-1])#id
    product_name.append(div.find("a").string)#name
    stock.append(div.find("div", {"class": "inventory"}).string.split(" ")[0])#stock
#     product_info[]
    
    
```

```python
location_ids=[]
location=[]
with open("/home/shreyas_ashwath/Desktop/a2-master/relations.csv","r") as r:
    reader=csv.reader(r)
    for rows in reader:
        location.append(rows)
    
    
    
for product_id in product_ids:
    for rows in location:
        if product_id==rows[0]:
            location_ids.append(rows[1])
            
```

```python
number,street,city,state=list(),list(),list(),list()
complete_location=[]
with open("/home/shreyas_ashwath/Desktop/a2-master/locations.csv","r") as re:
    reader1=csv.reader(re)
    for rows in reader1:
        complete_location.append(rows)
        
complete_details=[]        
# for i in range(len(product_name)):
for idx,location_id in enumerate(location_ids):
    for loc in complete_location:
        if location_id==loc[0]:
            complete_details.append([cost[idx],product_ids[idx],product_name[idx],stock[idx],loc[1],loc[2],loc[3],loc[4]])
```

```python
complete_details#[cost,product_ids,product_name,stock,number,street,city,state
```

```python
csv_rows=list()
for address in complete_details:
    cost,product_ids,product_name,stock,number,street,city,state=address
    addr=number+" "+street+" "+city+" "+state
    csv_rows.append([product_name,cost,stock,addr])

        
with open('final_output.csv', 'w') as csvFile:
    writer = csv.writer(csvFile)
    writer.writerows(csv_rows)

csvFile.close()   
```
